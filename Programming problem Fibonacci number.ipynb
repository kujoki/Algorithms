```python
import numpy as np 

def fib(n):
    arr = np.zeros((n,), dtype=int)
    if n == 1:
        return 1
    else:
        arr[0] = 1
        arr[1] = 1
        for _ in range(2,n):
            arr[_] = arr[_-1] + arr[_-2]
        return arr[n-1]

def main():
    n = int(input())
    print(fib(n))

if __name__ == "__main__":
    main()
```

```python
def fib_digit(n):
    a = 1
    b = 1
    for i in range(n-2):
        c = (a + b)%10
        a = b
        b = c
    return c


def main():
    n = int(input())
    print(fib_digit(n))


if __name__ == "__main__":
    main()
```

```python
def fib_mod(n, m):
    S = [0,1]

    for i in range(2, n+1):
        p = (S[i - 1] + S[i - 2]) % m #      m
        if p == 1 and S[i - 1] == 0: #  
            p = S[n % (i - 1)]
            break
        S.append(p)
    return p


def main():
    n, m = map(int, input().split())
    print(fib_mod(n, m))


if __name__ == "__main__":
    main()
    
```
